# Using Pandas Library To Read CSV Files

```python
import pandas as  pd
import numpy as np
import matplotlib.pyplot as plt
data = pd.read_csv('happyscore_income.csv')
data
#Printing as print(data) will give a different format, a not as a spreadsheet as seen below
```

# Pulling Out Columns

```python
happy = data["happyScore"]  #Or you could data.happyScore
A_income = data['avg_income']  #Or data.avg_income
print(happy)
print(A_income)
```

# Sorting The Data

```python
import pandas as pd
data = pd.DataFrame(data)
data.sort_values('avg_income', ascending = True, inplace = True)  #Inplace means the data itself is going to be affected and 
                                                                            #not just not just get a copy of it , p.s.False is by default
data
```

# Plotting avg_income against happyScore 

```python
plt.scatter(A_income,happy)
plt.xlabel('A_income')
plt.ylabel('Happy')
plt.show()
```

# Using Filters

```python
#Outputting countries with income higher than 15000
richest = data[data["avg_income"] >15000]
richest
#Outputting the country with lowest avg income above 15000
richest.iloc[0]

#With highest income
richest.iloc[-1] 

#Specifying a particular range
mean = np.mean(richest[0:5])
mean

#Comparing the average mean for the richest and all of them
richest_mean = np.mean(richest['avg_income'])
richest_mean

all_mean = np.mean(data['avg_income'])
all_mean

if richest_mean > all_mean:
    print('Richest mean ' ,richest_mean, 'is greatest')
else:
    print('All mean',all_mean, ' is greatest')
```

```python
#Plotting a graph for the richest countries avg. income against happyscore
plt.scatter(richest['avg_income'],richest['happyScore'])
#plt.show(), don't include it, will disrupt the name of the country e.e Italy and Luxemburg

richest.iloc[0]['avg_income']  
richest.iloc[0]['happyScore']  
plt.text(richest.iloc[0]['avg_income'],#Will give the avg income for the lowest above 15000
         richest.iloc[0]['happyScore'],#Will give happiest score for the country for the above country, both indexes are 0
         richest.iloc[0]['country']
        )
plt.text(richest.iloc[-1]['avg_income'],#Will give the avg income for the highest above 15000
         richest.iloc[-1]['happyScore'],#Will give happiest score for the country for the above country, both indexes are -1
         richest.iloc[-1]['country']
        )
```

# Using K-Means To Interpret Data

```python
from sklearn.cluster import KMeans
import numpy as np

data = pd.read_csv('happyscore_income.csv')

income = data['avg_income']
happy = data['happyScore']
income_happy = np.column_stack((income,happy))  #column_stack() takes 1 positional argument, ensure double brackets
income_happy

km_result = KMeans(n_clusters=3).fit(income_happy)
```

# To Find Cluster Centers

```python
km_result.cluster_centers_
```

# Plotting Cluster Centers In Graph

```python
cluster = km_result.cluster_centers_ #Renaming to a shorter name called cluster
plt.scatter(income,happy)
plt.scatter(cluster[:,0],cluster[:,1],s=1000,alpha=.70) #Size parameter to 1000
```

```python
from sklearn.cluster import KMeans
import numpy as np
X = np.array([[1, 2], [1, 4], [1, 0],
              [10, 2], [10, 4], [10, 0]])
kmeans = KMeans(n_clusters=2, random_state=0).fit(X)
kmeans.labels_

kmeans.predict([[0, 0], [12, 3]])

kmeans.cluster_centers_
```

```python
X = np.array([[1, 2], [1, 4], [1, 0],[10, 2], [10, 4], [10, 0]])
cluster1 = kmeans.cluster_centers_

plt.scatter(X[:,0],X[:,1])
plt.scatter(cluster1[:,0],cluster1[:,1],s= 1000,alpha= .25)
```
